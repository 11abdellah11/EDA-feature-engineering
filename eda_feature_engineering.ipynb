Dealing with Missing values

```python
# Missing val

feat_nan = [feat for feat in dataset.columns if dataset[feat].isnull().sum()>1]

for featur in feat_nan:
    print("the feature ", featur, 'and mission values is:', dataset[featur].isnull().mean(),'%')
```

Dealing with categorical feature

```python
categ_feat = [feat for feat in dataset.columns if dataset[feat].dtypes == 'O']
```

```python
categ_feat
```

```python
dataset[categ_feat[3]].dtype
```

```python
data = dataset.copy()
def replac_cat_by_missing(dataset,categorical_features):
    data[categorical_features] = dataset[categorical_features].fillna('Missing')
    return data
```

```python
replac_cat_by_missing(data,categ_feat)
```

```python
numerical_features_nan=[feature for feature in dataset.columns if dataset[feature].dtypes!='O' and
                       dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
for feat in numerical_features_nan:
    # we replace outliers by the median
    median_values = dataset[feat].median()
    #creat new featur
    dataset[feat+'_nan'] = np.where(dataset[feat].isnull(),1,0) 
    dataset[feat].fillna(median_values,inplace=True)
```
